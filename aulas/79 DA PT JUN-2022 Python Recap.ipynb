## Operadores
Operadores so operaes definidas entre 2 objetos. Quando estes objetos so nmeros (`int` e `float`), os operadores so as operaes matemticas. No entanto, operadores podem ser utilizados entre outros tipos de objetos.

Primeiro veremos a utilizao das operaes matemticas tradicionais:

```python
soma = 2 + 2
sub = 2 - 2
mult = 2 * 2
div = 2 / 3
exp = 2 ** 3
mod = 5%3
print(soma)
print(sub)
print(mult)
print(div)
print(exp)
print(mod)
```

Podemos utilizar os operadores de adio (entre dois `str`) e multiplicao em strings (entre um `str` e um `int`):

```python
soma = 'Pedro' + 'Teche'
mult = 'Pedro' * 5
print(soma)
print(mult)
```

e tambm em listas (adio entre duas `lists` e multiplicao entre uma `list` e um `int`)

```python
soma = [1, 2] + [3, 4]
mult = [1, 2] * 5
print(soma)
print(mult)
```

Podemos utilizar todos os operadores matemticos em `bools`: o valor `True`  convertido no `int` **1** e o valor `False` e convertido no `int` **0**.

```python
soma = True + True
sub = True + True - False
mult = True * True * True * True * False
div = 2 / True
print(soma)
print(sub)
print(mult)
print(div)
```

## Estruturas de Controle

Estruturas de controle nos permitem alterar *o fluxo do nosso cdigo*: executar certos comandos (linhas de cdigo) apenas quando certas condies so verdadeiras, ou ento repetir algumas linhas de cdigo de forma controlada. Para isto precisamos utilizar as 3 estruturas bsicas de controle: **if** (condicionais), **for** (loops) e **while** (loops condicionais).

### Condicionais (if, elif, else)
As estruturas condicionais nos permitem controlar a execuo das linhas a partir de condies: relaes de equivalncia, `==`, ou de comparao, `<`. Ao longo do curso veremos outras formas mais complexas de construir condies.

```python
1 == 2
```

```python
1 < 10
```

```python
if 1 == 2:
    print('Errado')
else:
    print('Certo')
```

```python
if 1 == 1:
    print('Certo')
else:
    print('Errado')
```

```python
x = 1
if x == 1:
    print('X  1')
else:
    print('X no  1')
```

A primeira condio  a que segue imediatamente aps o `if`. J o `else` nos permite executar certas linhas quando essa condio no  verdadeira.

Podemos utilizar o `elif` (lido *else if*) para encadear diversas condies em srie:

```python
x = input('Digite um nome')
if x == 'Pedro':
    print('Eu')
elif x == 'Adriano':
    print('Querido TA')
else:
    print('No sei...')
```

Alm de condies simples, podemos construir condies mais complexas atravs dos operadores lgicos `and` e `or`:

```python
0 < 10 and 0 > -5
```

```python
1 < 7 and 6 < 5
```

```python
1 < 7 or 6 < 5
```

```python
x = 1
if x >= 0 and x < 10:
    print('X entre 0 e 9')
elif x >= 10 and x < 20:
    print('X entre 10 e 19')
elif x < 0:
    print('X menor que 0')
```

### Loop (for)

Para executar um conjunto de linhas de forma repetida, podemos utilizar um **loop**. O loop  uma estrutura de controle que nos permite percorrer um **itervel** (por exemplo, uma lista) e executar um conjunto de linhas para cada elemento deste itervel.

```python
esta_lista = [1, 2, 3]
for elemento in esta_lista:
    elemento_quadrado = elemento ** 2
    print(elemento_quadrado)
```

Podemos utilizar condicionais dentro de loops:

```python
outra_lista = [1, 2, 3, '4', 'Pedro']
for elemento in outra_lista:
    if type(elemento) == int or type(elemento) == float:
        print(f'Elemento ao quadrado = {elemento ** 2}')
    else:
        print(f'Elemento original = {elemento}')
```

Por fim podemos utilizar loops dentro de loops (nesting):

```python
terceira_lista = [[1, 2, 3], [2, 3, 4], [4, 5, 6]]
for lista in terceira_lista:
    print(lista)
    for elemento in lista:
        print(f'Elemento ao quadrado = {elemento ** 2}')
```

### Loop condicional (while)

Para executar um conjunto de linhas de forma repetida, podemos utilizar um **loop**. O loop  uma estrutura de controle que nos permite percorrer um **itervel** (por exemplo, uma lista) e executar um conjunto de linhas para cada elemento deste itervel.

```python
x = 0
while x < 10:
    x = x + 1
    print(x)
```

```python
x = None
while x != 'Pedro':
    x = input('Digite meu nome: ')
    if x == 'Pedro':
        print('Parabns!')
    else:
        print('Tente de novo...')
```

Alm dos dois exemplos acima, uma construo muito comum em loops condicionais so os *loops infinitos* (`while True:`). Esta forma nos permite construir condies de quebra complexas dentro do loop atravs de condicionais normas e do comando `break`.

```python
while True:
    nome = input('Digite um nome: ')
    if nome == 'Adriano' or nome == 'Pedro':
        print('Parabns!')
        break
    else:
        print('Tente de novo...')
```

# Desafio

Conforme vimos anteriormente, listas podem conter listas. E as listas dentro de uma lista podem conter mais listas ainda... Esse tipo de estrutura  muito comum. O nmero de listas dentro de listas  chamado de **profundidade**. Por exemplo, a lista abaixo tem profundidade 2 (seu terceiro elemento  uma lista simples):

```python
p2 = [1, 2, [1, 2, 3]]
print(p2)
```

Enquanto a prxima tem profundidade 3 (seu terceiro elemento  uma lista que tambm contm uma lista):

```python
p3 = [1, 2, [1, 2, [1, 2]]]
print(p3)
```

Uma tarefa comum  *desempacotar* uma lista: sair de `[1, [1, 2]]` para `[1, 1, 2]`.
