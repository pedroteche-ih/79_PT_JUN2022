```python
import pandas as pd
import numpy as np
import seaborn as sns
import patsy
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
from sklearn.feature_selection import RFECV
```

# Limpeza de dados

Limpeza de 0s nas dimens천es

```python
diamonds = pd.read_csv('tb_diamantes.csv')
x_0 = diamonds['x'] == 0
y_0 = diamonds['y'] == 0
z_0 = diamonds['z'] == 0
dim_0 = x_0 | y_0 | z_0
diamonds_c = diamonds[~dim_0].copy()
print('Linhas na etapa atual:', str(len(diamonds_c)))
```

```python
dict_clarity = {
    'I1': 0,
    'SI2': 1,
    'SI1': 2,
    'VS2': 3,
    'VS1': 4,
    'VVS2': 5,
    'VVS1': 6,
    'IF': 7
}
diamonds_c['clarity_num'] = diamonds_c['clarity'].map(dict_clarity)
```

```python
dict_color = {
    'D': 0,
    'E': 1,
    'F': 2,
    'G': 3,
    'H': 4,
    'I': 5,
    'J': 6,
}
diamonds_c['color_num'] = diamonds_c['color'].map(dict_color)
```

```python
dict_cut = {
    'Fair': 0,
    'Good': 1,
    'Very Good': 2,
    'Premium': 3,
    'Ideal': 4,
}
diamonds_c['cut_num'] = diamonds_c['cut'].map(dict_cut)
```

Limpeza de outliers

```python
diamonds_c['volume'] = diamonds_c['x'] * diamonds_c['y'] * diamonds_c['z']
diamonds_c['price_k'] = diamonds_c['price']/diamonds_c['carat']
diamonds_c['density'] = diamonds_c['carat']/diamonds_c['volume']
```

```python
pricek_inf = np.quantile(diamonds_c['price_k'], 0.01)
pricek_sup = np.quantile(diamonds_c['price_k'], 0.99)
density_inf = np.quantile(diamonds_c['density'], 0.01)
density_sup = np.quantile(diamonds_c['density'], 0.99)

pricek_in = (pricek_inf < diamonds_c['price_k']) &  (diamonds_c['price_k'] < pricek_sup)
density_in = (density_inf < diamonds_c['density']) &  (diamonds_c['density'] < density_sup)
diamonds_c = diamonds_c[pricek_in & density_in].copy()
print('Linhas na etapa atual:', str(len(diamonds_c)))
```

```python
diamonds_c.info()
```

# Modelo I

```python
string_x = ' ~ log_carat * color_num * clarity_num'
Y, X = patsy.dmatrices('log_price' + string_x, diamonds_c)
modelo = LinearRegression()
lm_fit = modelo.fit(X, Y)
diamonds_c['pred_price'] = np.exp(lm_fit.predict(X))
```

```python
sns.scatterplot(data = diamonds_c, x = 'pred_price', y = 'price')
```

```python
tb_rick = pd.read_csv('tb_rick_diamantes.csv')
tb_rick['log_carat'] = np.log(tb_rick['carat'])
tb_rick['std_dim'] = tb_rick[['x', 'y', 'z']].std(axis = 1, ddof=0)/(tb_rick[['x', 'y', 'z']].sum(axis = 1))
tb_rick['clarity_num'] = tb_rick['clarity'].map(dict_clarity)
tb_rick['color_num'] = tb_rick['color'].map(dict_color)
X_rick = patsy.dmatrix(string_x, tb_rick)
tb_rick['price_predicted'] = np.exp(lm_fit.predict(X_rick))
tb_rick.to_csv('price_predicted.csv')
```

# Modelo IV

```python
string_x = '~ log_carat * color_num * clarity_num * cut * depth * max_dim_classif + mudim_carat : mudim_carat_classif'
Y, X = patsy.dmatrices('log_price' + string_x, diamonds_c)
modelo = LinearRegression()
lm_fit = modelo.fit(X, Y)
```

```python
tb_rick = pd.read_csv('tb_rick_diamantes.csv')
tb_rick['log_carat'] = np.log(tb_rick['carat'])
tb_rick['std_dim'] = tb_rick[['x', 'y', 'z']].std(axis = 1, ddof=0)/(tb_rick[['x', 'y', 'z']].sum(axis = 1))
tb_rick['clarity_num'] = tb_rick['clarity'].map(dict_clarity)
tb_rick['color_num'] = tb_rick['color'].map(dict_color)
tb_rick['max_dim'] = tb_rick[['x', 'y', 'z']].max(axis = 1)
tb_rick['max_dim_classif'] = np.where(tb_rick['max_dim'] > 8, 'GRANDE', 'PEQUENO')
tb_rick['mean_dim'] = tb_rick[['x', 'y', 'z']].mean(axis = 1)
tb_rick['mudim_carat'] = tb_rick['mean_dim']/tb_rick['carat']
tb_rick.loc[:, 'mudim_carat_classif'] = 'PEQUENO'
tb_rick.loc[tb_rick['mudim_carat'] > 4, 'mudim_carat_classif'] = '>4'
X_rick = patsy.dmatrix(string_x, tb_rick)
tb_rick['price_predicted'] = np.exp(lm_fit.predict(X_rick))
tb_rick.to_csv('price_predicted.csv')
```

RMSE = 732.6496515918401
